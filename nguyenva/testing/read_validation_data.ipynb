```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
raw = pd.read_excel('../../raw_data/Validation/20190610_VN.xlsx', sheet_name = 'Sheet1')
```

```python
# Rename columns
samples = {
    'W298H' : ('A', np.arange(1, 7)),
    'T293D' : ('B', np.arange(1, 7)),
    'G291L' : ('C', np.arange(1, 7)),
    'T443K' : ('D', np.arange(1, 7)),
    'F512A' : ('E', np.arange(1, 7)),
    'F512P' : ('F', np.arange(1, 7)),
    'K298M' : ('G', np.arange(1, 7)),
    'WT' : ('H', np.arange(1, 7)),
    'Empty' : ('A', np.arange(7, 13))
}

sample_rename_dict = {}
for sample in list(samples.keys()):
    row, cols = samples[sample]
    rad_cols = cols[:3]
    dmso_cols = cols[3:]
    
    for idx, col in enumerate(rad_cols):
        sample_name = sample + '_rad_' + str(idx + 1)
        col_name = row + str(col)
        sample_rename_dict[col_name] = sample_name
        
    for idx, col in enumerate(dmso_cols):
        sample_name = sample + '_dmso_' + str(idx + 1)
        col_name = row + str(col)
        sample_rename_dict[col_name] = sample_name
        
raw.rename(columns=sample_rename_dict, inplace=True)
```

```python
# Save renamed dataframe
raw.to_csv('../../raw_data/Validation/20190610_renamed.csv', index=False)
```

```python
# Plot curves
time = np.linspace(0.5, 72.5, len(raw))
time_x = np.concatenate(
    (
        time,
        time,
        time
    )
)
```

```python
idx = 0
fig, ax = plt.subplots(3, 3, figsize=(10,10))
for sample in list(samples.keys()):
    for condition in ['rad', 'dmso']:
        sample_names = [
            '{}_{}_{}'.format(sample, condition, 1),
            '{}_{}_{}'.format(sample, condition, 2),
            '{}_{}_{}'.format(sample, condition, 3),
        ]
        y = np.concatenate(
            (
                raw[sample_names[0]].values,
                raw[sample_names[1]].values,
                raw[sample_names[2]].values,
            )
        )
        long_temp_df = pd.DataFrame((time_x, y)).T
        long_temp_df.rename(columns={0 : 'time', 1: '{}_{}'.format(sample, condition)}, inplace=True)
        row_idx = int(idx / 3)
        col_idx = idx % 3
        sns.lineplot(x='time', y='{}_{}'.format(sample, condition), data=long_temp_df, ci='sd', ax=ax[row_idx, col_idx])
    plt.legend()
    idx += 1
    
    
    
    
y = np.concatenate(
    (
        raw['W298H_rad_1'].values,
        raw['W298H_rad_2'].values,
        raw['W298H_rad_3'].values
    )
)
long_temp_df = pd.DataFrame((time_x, y)).T
long_temp_df.rename(columns={0 : 'time', 1: 'rep1'}, inplace=True)
```

```python
growth_rates = []
conditions = []
sample_name_order = []
n_min = 72 * 60
for sample in list(samples.keys()):
    for condition in ['rad', 'dmso']:
        sample_names = [
            '{}_{}_{}'.format(sample, condition, 1),
            '{}_{}_{}'.format(sample, condition, 2),
            '{}_{}_{}'.format(sample, condition, 3),
        ]
        for idx in np.arange(3):
            vals = raw[sample_names[idx]].values
            rate = (vals[-1] - vals[0]) / n_min
            
            # Append growth rate, sample names, and condition
            growth_rates.append(rate)
            sample_name_order.append(sample)
            conditions.append(condition)
        
```

```python
sns.barplot(x=sample_name_order, y=growth_rates, hue=conditions)
plt.xlabel('Variant')
plt.ylabel('OD600/min')
```

```python
# Grab activity scores
point_muts = pd.read_csv('../../utils/point_muts.csv', index_col=0, header=[0,1])
resids = pd.read_csv('../../utils/residuals.csv', index_col=0)

point_muts.rename(index=dict(zip(point_muts.index.tolist(), resids.index.tolist())), inplace=True)
```

```python
variants = [
    'W289H',
    'T293D',
    'G291L',
    'T443K', 
    'F512A',
    'F512P',
    'K298M',
    '_wt'
]

bar_width=0.45
rad_scores = []
dmso_scores = []
rad_errs = []
dmso_errs = []

# Set position of bar on X axis
r1 = np.arange(len(variants))
r2 = [x + bar_width for x in r1]

for variant in variants:
    rad_score = point_muts.loc[variant, ('Radicicol', 'score')]
    rad_err = point_muts.loc[variant, ('Radicicol', 'SE')]
    dmso_score = point_muts.loc[variant, ('Zero', 'score')]
    dmso_err = point_muts.loc[variant, ('Zero', 'SE')]
    
    # Append to lists
    rad_scores.append(rad_score)
    rad_errs.append(rad_err)
    dmso_scores.append(dmso_score)
    dmso_errs.append(dmso_err)
```

```python
fig, ax = plt.subplots()
ax.bar(r1, rad_scores, yerr=rad_errs, width=bar_width, label='rad')
ax.bar(r2, dmso_scores, yerr=dmso_errs, width=bar_width, label='dmso')
ax.legend()
ax.set_xticks(r1)
ax.set_xticklabels(variants)
ax.set_xlabel('Variant')
ax.set_ylabel('Enrich2 activity score')
```
