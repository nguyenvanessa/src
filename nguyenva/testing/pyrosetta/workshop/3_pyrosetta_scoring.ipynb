```python
from pyrosetta import *
pyrosetta.init()
```

```python
# Load pose from PDB directly (if you have an internet connection)
from pyrosetta.toolbox import pose_from_rcsb
ras = pose_from_rcsb("6Q21")
print(ras)
```

```python
# Cleaned 6Q21.clean.pdb in vim
# Reload
ras = pose_from_pdb("6Q21.clean.pdb")
print(ras.pdb_info())
```

```python
# 1. Define a scoring function
from pyrosetta.teaching import *
scorefxn = get_fa_scorefxn()  # Default full atom score function
```

```python
print(scorefxn)
```

```python
# You can also set up your own scorefunction
scorefxn2 = ScoreFunction()
scorefxn2.set_weight(fa_atr, 1.0)
scorefxn2.set_weight(fa_rep, 1.0)
```

```python
# Evaluate energy of Ras with standard score function
print(scorefxn(ras))
```

```python
# Breakdown energy into individual pieces with show method
scorefxn.show(ras)

# 3 most dominant contributions:
# fa_atr (-1039)
# fa_dun (907)
# fa_sol (682)
```

```python
# Unweighted, individual component energies of
# each residue are stored in Pose and can be
# accessed by its energies() object

print(ras.energies().show(24)) # 24th residue
```

```python
# VDW and solvation terms are atom-atom pairwise
# energies calculated from a lookup table 
# dependent between two atoms and their types
# You can access the table (etable) directly

r1 = ras.residue(24)
r2 = ras.residue(20)

a1 = r1.atom("N")
a2 = r2.atom("O")

r1_idx = r1.atom_index("N")
r2_idx = r2.atom_index("O")

etable_atom_pair_energies(r1, r1_idx, r2, r2_idx, scorefxn)

# 
```

```python
# Hydrogen bonding energies are stored in an HBondSet
from rosetta.core.scoring.hbonds import HBondSet

# Create HBondSet object
hbond_set = HBondSet()

# Fill set from pose (after making sure the pose
# has had its Energies object updated based on
# neighboring residues within the pose)
ras.update_residue_neighbors()
hbonds.fill_hbond_set(ras, False, hbond_set)
hbond_set.show(ras)
```

```python
# Above steps have been combined into get_hbonds
# methods
# This is not available in my mini PyRosetta
hbond_set.show(ras, 24)
```

```python
# Analyze energy between residues Y102 and Q408
# in cetuximab

# 1. Load pose from PDB directly (if you have an internet connection)
cet = pose_from_rcsb("1YY9")

# 2. Get residue index of D:Y102 and A:Q408
y102_res_idx = cet.pdb_info().pdb2pose('D', 102)
q408_res_idx = cet.pdb_info().pdb2pose('A', 408)

# 3. Create emap
emap = EMapVector()

# 4. Create Residue objects from residue indices
rsd1 = cet.residue(y102_res_idx)
rsd2 = cet.residue(q408_res_idx)

# 5. Score Residue objects, pose, and emap
scorefxn.eval_ci_2b(rsd1, rsd2, cet, emap)

# 6. Print VDW and solvation energies
print(emap[fa_atr])
print(emap[fa_rep])
print(emap[fa_sol])
```
