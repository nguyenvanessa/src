## Basic PyRosetta

```python
# Clean PDB file
from pyrosetta.toolbox import cleanATOM
cleanATOM("1YY8.pdb")  # Creates 1YY8.clean.pdb in the same directory
```

```python
# Load protein pose
from pyrosetta import pose_from_pdb
pose = pose_from_pdb("1YY8.clean.pdb")  # Creates Pose object
```

```python
# Load pose from PDB directly (if you have an internet connection)
from pyrosetta.toolbox import pose_from_rcsb
pose = pose_from_rcsb("1YY8")
```

```python
## Examine protein with a variety of queries
print(pose)
print(pose.sequence())
print("Protein has", pose.total_residue(), "residues")
print(pose.residue(500).name())
```

```python
print(pose.pdb_info().chain(500))
print(pose.pdb_info().number(500))
# NOTE: the 500th residue in the pdb file is not necessarily the
# 500th residue in the protein
```

```python
# Rosetta has internal numbering that you can look up
print(pose.pdb_info().pdb2pose('B', 100))

# The converse command is
print(pose.pdb_info().pose2pdb(25))
```

## Protein geometry

```python
# Find phi, psi, chi_1 dihedral angles of residue 5
print(pose.phi(5))
print(pose.psi(5))
print(pose.chi(1, 5))
```

```python
# Find N-Ca and Ca-C bond lengths

# 1. Store unique atom identifier codes in variables
R5N = AtomID(1, 5)
R5CA = AtomID(2, 5)
R5C = AtomID(3, 5)
# NOTE: this works because the atoms are listed in a
# consistent order in a pdb file

# 2. Use identifier codes to look up bond lengths
print(pose.conformation().bond_length(R5N, R5CA))
print(pose.conformation().bond_length(R5CA, R5C))
```

```python
# ALTERNATE method for finding bond lengths
N_xyz = pose.residue(5).xyz("N")
CA_xyz = pose.residue(5).xyz("CA")
N_CA_vector = CA_xyz - N_xyz
print(N_CA_vector.norm())
```
