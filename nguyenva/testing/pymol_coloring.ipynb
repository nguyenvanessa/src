# 1. Check if the numbering from the DMS matches the 2src_kinDom.pdb numbering

```python
master = pd.read_csv('../../utils/point_muts.csv', index_col=0, header=[0,1])
```

```python
aa_key = {
    'His' : 'H',
    'Lys' : 'K',
    'Arg' : 'R',
    'Asp' : 'D',
    'Glu' : 'E',
    'Cys' : 'C',
    'Met' : 'M',
    'Asn' : 'N',
    'Gln' : 'Q',
    'Ser' : 'S',
    'Thr' : 'T',
    'Ala' : 'A',
    'Ile' : 'I',
    'Leu' : 'L',
    'Val' : 'V',
    'Phe' : 'F',
    'Trp' : 'W',
    'Tyr' : 'Y',
    'Gly' : 'G',
    'Pro' : 'P',
    'Ter' : '*'
    
}

wt_seq = []
for variant in master.index.tolist():
    
    # Skip wt and sy rows
    if 'p.' not in variant:
        continue
        
    # Grab wt_aa and DMS position
    wt_aa = variant[2:5]
    pos = variant[5:-3]
    
    # Append to list
    pos_aa = pos + aa_key[wt_aa]
    wt_seq.append(pos_aa)

# Create a sorted set from variant list
wt_seq = list(set(wt_seq))
wt_seq.sort()

# Separate into a list of positions and a list of AAs
aa_seq = []
pos_seq = []

for pos_aa in wt_seq:
    aa_seq.append(pos_aa[-1])
    pos_seq.append(pos_aa[:-1])
    
```
