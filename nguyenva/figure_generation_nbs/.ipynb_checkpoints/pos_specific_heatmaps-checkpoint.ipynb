```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import DivergingNorm
import sys
sys.path.insert(1, '../../utils/')
import src_utils
```

# 1. Define position clusters


```python
clust1 = [274, 276, 283, 285, 286]  # G-loop
clust2 = [303, 335]  # aC helix hbond
clust3 = [323, 343, 399, 400, 404]  # SH3-SH2 linker interface with kinase domain
clust4 = [381, 444, 512]  # aF pocket
```

```python
'+'.join([str(x) for x in clust4])
```

# 2. Read in heatmaps

```python
diffsel = pd.read_csv('../heatmaps/envision_diffsel_heatmap.csv', index_col=0).values
dmso = pd.read_csv('../heatmaps/dmso_heatmap.csv', index_col=0).values
rad = pd.read_csv('../heatmaps/rad_heatmap.csv', index_col=0).values
```

# 3. Plot heatmaps for each cluster

```python
max_abs = np.nanmax(abs(np.concatenate((diffsel, dmso, rad), axis=-1)).flatten())
```

## G loop

```python
# Define cluster and idx and title
cluster = clust1
idx = [src_utils.__pos_idx_dict__[pos] for pos in cluster]
title = 'G-loop'

# Create figure and axes objects
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(10,5))

# Plot individual heatmaps
heatmap = dmso[idx, :]
hm1 = axes[0].imshow(heatmap.T, cmap='bwr', vmin=-max_abs, vmax=max_abs)

heatmap = rad[idx, :]
hm2 = axes[1].imshow(heatmap.T, cmap='bwr', vmin=-max_abs, vmax=max_abs)

heatmap = diffsel[idx, :]
hm3 = axes[2].imshow(heatmap.T, cmap='bwr', vmin=-max_abs, vmax=max_abs)

# Adjust tick labels and positions
tick_labels = cluster
for ax in [axes[0], axes[1], axes[2]]:
    
    # Set tick locations
    ax.set_yticks(np.arange(heatmap.shape[1]))
    ax.set_xticks(np.arange(heatmap.shape[0]))

    # Set tick labels
    ax.set_yticklabels(src_utils.__aa_idx_dict__.keys())
    ax.set_xticklabels(tick_labels, rotation=90)

# Set colorbar to be identical for each heatmap
fig.colorbar(hm1, cmap='bwr', ax=axes, shrink=0.5)

# Set axes labels
axes[0].set(ylabel='DMSO')
axes[1].set(ylabel='Radicicol')
axes[2].set(ylabel='Diffsel')

# Set title
fig.suptitle(title)

# Show figure
plt.show()
```

## aC helix hbond

```python
# Define cluster and idx and title
cluster = clust2
idx = [src_utils.__pos_idx_dict__[pos] for pos in cluster]
title = 'aC helix hbond'

# Create figure and axes objects
fig, axes = plt.subplots(nrows=3, ncols=1, figsize=(5, 12.5))

# Plot individual heatmaps
heatmap = dmso[idx, :]
hm1 = axes[0].imshow(heatmap.T, cmap='bwr', vmin=-max_abs, vmax=max_abs)

heatmap = rad[idx, :]
hm2 = axes[1].imshow(heatmap.T, cmap='bwr', vmin=-max_abs, vmax=max_abs)

heatmap = diffsel[idx, :]
hm3 = axes[2].imshow(heatmap.T, cmap='bwr', vmin=-max_abs, vmax=max_abs)

# Adjust tick labels and positions
tick_labels = cluster
for ax in [axes[0], axes[1], axes[2]]:
    
    # Set tick locations
    ax.set_yticks(np.arange(heatmap.shape[1]))
    ax.set_xticks(np.arange(heatmap.shape[0]))

    # Set tick labels
    ax.set_yticklabels(src_utils.__aa_idx_dict__.keys())
    ax.set_xticklabels(tick_labels, rotation=90)

# Set colorbar to be identical for each heatmap
fig.colorbar(hm1, cmap='bwr', ax=axes, shrink=0.5)

# Set axes labels
axes[0].set(ylabel='DMSO')
axes[1].set(ylabel='Radicicol')
axes[2].set(ylabel='Diffsel')

# Set title
fig.suptitle(title)

# Show figure
plt.show()
```
