```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
raw = pd.read_excel('20190513_VN.xlsx', sheet_name = 'Sheet1')
```

```python
rows = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H']
cols = np.arange(1, 13)

conditions = ['radicicol', 'DMSO']

samples = ['K298M', 'WT', 'Empty', 'W289H', 'T293D', 'G291L', 'T443K', 'F512A', 'F512P']
#samples = ['HSP1', 'HSP2', 'HSP3', 'HSP4', 'HSP5', 'HSP6', 'HSP7', 'HSP8' ,'HSP9']
wellsPerSample = 6
sampleCount = 9

colLabels = []
wellID = []
sampleLabels = []

row = 0
col = wellsPerSample
for idx in np.arange(9):
    if row > 7:
        row = row % 7 - 1
        col = col + 6
    colLabels = colLabels + [rows[row] + str(i) for i in np.arange(col - 5, col + 1)]
    row += 1
    
    sampleLabels = sampleLabels + [conditions[0] + '_' + str(idx) for num in np.arange(1, 4)]
    sampleLabels = sampleLabels + [conditions[1] + '_' + str(idx) for num in np.arange(1, 4)]

raw.rename(dict(zip(colLabels, sampleLabels)), axis = 1, inplace = True)

idx = ['_' not in name for name in raw.columns]
raw.drop(raw.columns[idx][2:], axis = 1, inplace = True)

raw['id'] = raw.index
```
